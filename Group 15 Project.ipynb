```python
import numpy as np
import pandas as pd
```

```python
DataWorld = pd.read_csv("Crops_AllData_Normalized.csv", encoding = 'cp1252')
```

```python
DataWorld
```

### Drop the unneeded columns

```python
DataWorld.drop(["Area Code", "Item Code", "Element Code", "Year Code", "Flag"], inplace = True, axis = 1)
```

```python
DataWorld
```

### Filter the data to show only the production value of each crop

```python
DataWorld_Production = DataWorld.loc[DataWorld.Element == "Production"]
```

```python
DataWorld_Production
```

### Filter the data to show only data on crops in Nigeria


```python
DataWorld_Nig = DataWorld_Production.loc[DataWorld.Area == "Nigeria"]
```

```python
DataWorld_Nig
```

### Find out which of the columns has missing value

```python
# Its being discovered that column "Value" has missing values

DataWorld_Nig.info()
```

### How many missing values are in column "Value"

```python
# column "Value" has 123 missing values

DataWorld_Nig.isnull().sum()
```

### Dropping the rows having missing values

```python
DataWorld_Nig.dropna(inplace = True)
```

```python
DataWorld_Nig
```

```python
# Removing rows with value = 0.0

DataWorld_Nigeria = DataWorld_Nig[DataWorld_Nig.apply(lambda x:x["Value"] != 0.0, axis = 1)]
```

```python
DataWorld_Nigeria
```

```python
# The missing/null valuesin column "Value" have being removed

DataWorld_Nigeria.isnull().sum()
```

### Resetting the index of the data

```python
DataWorld_Nigeria.reset_index(inplace = True)
```

```python
DataWorld_Nigeria
```

```python
DataWorld_Nigeria.to_csv("DataWorld_Nigeria.csv")
```

## Import the Population Data

```python
World_Pop = pd.read_csv("Total_Population_All_Countries.csv", encoding = "cp1252")
```

```python
World_Pop
```

### Filter the Population Data to reveal only the data for Nigeria

```python
World_Pop_Nig = World_Pop.loc[World_Pop.Location == "Nigeria"]
```

```python
World_Pop_Nig
```

## Drop Unneeded Columns


```python
Drop_Col = ["LocID","VarID","MidPeriod","PopMale","PopFemale"]

World_Pop_Nig.drop(Drop_Col, inplace = True, axis = 1)
```

```python
World_Pop_Nig
```

## Data between 1981 and 2019

```python
World_Pop_Nig1 = World_Pop_Nig.loc[(World_Pop_Nig.Time > 1980) & (World_Pop_Nig.Time < 2020)]
```

```python
World_Pop_Nig1.head()
```

### Resetting the index

```python
World_Pop_Nig1.reset_index(inplace = True)
```

```python
World_Pop_Nig1.head()
```

### Checking for null/missing Values

```python
# No missing/null values

World_Pop_Nig1.isnull().sum()
```

```python
World_Pop_Nig1.to_csv("Nig_Population.csv")
```
